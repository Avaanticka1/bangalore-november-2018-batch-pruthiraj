### 5. Calculating distortion - C distance

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

Iterating values of k from 1 to 10 fit K means model
Using c distance - Get the measure for Sum of squares error.

```python
cluster_range = range(2,10)
distortion = []
cluster_sil_score = []

for num_clusters in cluster_range:
    clusters = KMeans(n_clusters=num_clusters,init='k-means++',n_init=5)
    clusters.fit(car1)
    lables = clusters.labels_
    centroids = clusters.cluster_centers_
    distortion.append(clusters.inertia_)
    cluster_sil_score.append(silhouette_score(car1,labels=lables,metric='euclidean'))
```

### 6. Plotting Elbow/ Scree Plot

```python
cluster_df = pd.DataFrame({'num_of_cluster':cluster_range,'cluster_error':distortion,"avg_sil_score":cluster_sil_score})
```

```python
plt.figure(figsize=(15,5))
plt.plot(cluster_df.num_of_cluster,cluster_df.cluster_error,marker = 'o')
```

### Find out the optimal value of K

```python
cluster_real = KMeans(n_clusters=4,n_init=10)
cluster_real.fit(car1)
```

## Hierarchical Clustering 

### 8. Variable creation

For Hierarchical clustering, we will create datasets using multivariate normal distribution to visually observe how the clusters are formed at the end

```python
a = np.random.multivariate_normal([10, 0], [[3, 1], [1, 4]], size=[100,])
b = np.random.multivariate_normal([0, 20], [[3, 1], [1, 4]], size=[50,])
c = np.random.multivariate_normal([10, 20], [[3, 1], [1, 4]], size=[100,])
```

### 9. Combine all three arrays a,b,c into a dataframe

```python
na = np.concatenate((a,b,c),axis =0)
result = pd.DataFrame(na,columns=['A','B'])
```

```python
result.shape
result.size
```

### 10. Use scatter matrix to print all the 3 distributions

```python
sns.pairplot(data=result,diag_kind='kde')
```

### 11. Find out the linkage matrix

```python
from sklearn.cluster import AgglomerativeClustering 
```

```python
model = AgglomerativeClustering(n_clusters=3, affinity='euclidean',  linkage='ward')
model.fit(result)
```

```python
result['labels'] = model.labels_
```

```python
result.groupby('labels').count().plot(kind = 'bar')
plt.grid()
plt.show()
```

### 12. Plot the dendrogram for the consolidated dataframe

```python
from scipy.cluster.hierarchy import cophenet, dendrogram, linkage
from scipy.spatial.distance import pdist
```

```python
z = linkage(result,'average')
```

```python
c, coph_dist = cophenet(z,pdist(result))
```

### 13. Recreate the dendrogram for last 12 merged clusters 

```python
z.shape
```

```python
plt.figure(figsize=(10, 10))
plt.title('Dendogram')
plt.xlabel('index')
plt.ylabel('Distances')
a = dendrogram(z, leaf_rotation=90.,color_threshold = 10, leaf_font_size=8.)
plt.tight_layout()
```

### 14. From the truncated dendrogram, find out the optimal distance between clusters which u want to use an input for clustering data

```python
plt.figure(figsize=(10, 10))
plt.title('Dendogram')
plt.xlabel('index')
plt.ylabel('Distances')
a = dendrogram(z, leaf_rotation=90.,color_threshold = 8, leaf_font_size=8.,truncate_mode='lastp' )
plt.tight_layout()
```
