
### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
def ran(a,b):
    ran = a-b
    return ran

list1 = ['Sepal Length (in cm)', 'Sepal Width in (cm)','Petal length (in cm)','Petal width (in cm)']

for i in range(0,4):
    a = iris[list1[i]].max()
    b = iris[list1[i]].min()
    print(list1[i] ,"range =",ran(a,b))
```
