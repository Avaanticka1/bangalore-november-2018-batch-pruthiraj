###### Split the data into training and test set in the ratio of 70:30 respectively

```python
from sklearn.model_selection import train_test_split
X = bank_df.drop(columns=['Personal Loan'],axis = 1)
Y = bank_df.iloc[:,9]
train_set, test_set, train_labels, test_labels = train_test_split(X, Y, test_size=0.30, random_state=1)
```

###### Use different classification models (Logistic, Naïve Bayes) to predict the likelihood of a liability customer buying personal loans

```python
from sklearn import preprocessing
train_set_scaled = preprocessing.scale(train_set)
test_set_scaled = preprocessing.scale(test_set)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(train_set_scaled, train_labels)
```

```python
print("Training score {}",model.score(train_set_scaled,train_labels))
print("Testing score {}",model.score(test_set_scaled,test_labels))
```

```python
y_LRpredict = model.predict(test_set_scaled)
```

```python
from sklearn.naive_bayes import GaussianNB
nv_gnb_model = GaussianNB()
nv_gnb_model.fit(train_set_scaled,train_labels)
```

```python
print("Training score {}",nv_gnb_model.score(train_set_scaled,train_labels))
print("Testing score {}",nv_gnb_model.score(test_set_scaled,test_labels))
```

```python
y_NVpredict = nv_gnb_model.predict(test_set_scaled)
```

###### Print the confusion matrix for all the above models (5 points)

```python
from sklearn.metrics import confusion_matrix
from sklearn import metrics
print("Confusion Matrix for Logistic Regression")
metrics.confusion_matrix(test_labels,y_LRpredict)
```

```python
print("Confusion Matrix for Naïve Bayes")
metrics.confusion_matrix(test_labels,y_NVpredict)
```

###### Give your reasoning on which is the best model in this case and why it performs better? (5 points)

Logistic regression better performs here. This is clear from the test scores and confusion matrix of logistic regression.

Also the Naive bayes model works based on conditional probablity which is dependend on evidence that as occured. So Naive bayes model will better perform on the dataset which has more bias.
In the current dataset the attributes like income, CC Avergae, CD account have good corelation with personal loan, which make the Logistic regression directly apply the sigmoid function and get the probablity to predict. 
Due to above mention reasons the Logistic regression performs better for this dataset.
