# Reading Data and replacing "-" with 0

```python
ds = pd.read_csv("Laliga.csv")
ds.columns = ds.iloc[0,]
ds = ds.iloc[1:,:]
```

```python
ds.head()
```

```python
ds = ds.set_index(keys="Pos",drop = True)
ds =ds.replace(to_replace= "-",value=0)
```

```python
ds["GamesWon"] = pd.to_numeric(ds["GamesWon"])
ds["GamesPlayed"] = pd.to_numeric(ds["GamesPlayed"])
```

# Creating Winning Percentage column 

```python
ds["Winning Percent"] = round((ds["GamesWon"] / ds["GamesPlayed"])*100,2)
```

```python
ds["Winning Percent"].replace(np.nan,0,inplace=True)
```
